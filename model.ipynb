```python
import pandas as pd

# Load the real dataset
df = pd.read_csv("data.csv")

# Show the first few rows
df.head(10)
```

```python
#Visualization of the data

import matplotlib.pyplot as plt

#Scatter plot of the data
plt.scatter(df['Hours'], df['Scores'], color='blue')
plt.title('Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```

```python
#Train a Linear Regression model

from sklearn.linear_model import LinearRegression

#Step 1: Define input (X) and output (y)
X = df[['Hours']]
y = df['Scores']

#Step 2: Create and train the model
model = LinearRegression()
model.fit(X, y)

#Step 3: View the model parameters
print(f"Intercept (θ₀): {model.intercept_}")
print(f"Slope (θ₁): {model.coef_[0]}")


#Visualize the regression line

#Step 1: Make predictions using the model
predicted_scores = model.predict(X)

#Step 2: Plot actual data points
plt.scatter(X, y, color='blue', label='Actual Data')

# Step 3: Plot the regression line
plt.plot(X, predicted_scores, color='red', label='Regression Line')

# Step 4: Add titles and labels
plt.title('Hours vs Exam Score (Regression Line)')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Predict
# Predict score for 9.25 hours
hours = [[9.25]]
predicted_score = model.predict(hours)
print(f"Predicted Score for 9.25 hours of study: {predicted_score[0]:.2f}")
```

```python
#Evaluate the model
from sklearn.metrics import mean_squared_error, r2_score
# Step 1: Make predictions on the training set
y_pred = model.predict(X)
# Step 2: Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y, y_pred)
# Step 3: Calculate R-squared (R²)
r2 = r2_score(y, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared: {r2:.2f}")
```
